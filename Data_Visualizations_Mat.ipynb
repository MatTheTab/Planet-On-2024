```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def get_correlation(merged_df, price_label):
    data = merged_df[merged_df[price_label] != 0]
    data = data[data[price_label] < 1e7]
    correlation8_2 = data["contractors_y"].corr(data[price_label])
    # Plot the violin plot for "8.2.)" grouped by "7.5_combined"
    plt.figure(figsize=(8, 6))
    sns.violinplot(x="contractors_y", y=price_label, data=data, palette="muted", scale="width")

    # Add labels and title
    plt.title(f"Distribution of {price_label} Grouped by contractors_y", fontsize=14)
    plt.xlabel("contractors_y", fontsize=12)
    plt.ylabel(price_label, fontsize=12)

    # Show the plot
    plt.tight_layout()
    plt.show()

    return correlation8_2

sns.set_theme(style="whitegrid") 
correlation_6 = get_correlation(merged_df, "6.2.)")
correlation_8 = get_correlation(merged_df, "8.2.)")
correlation_431 = get_correlation(merged_df, "4.3.1)")

custom_green_palette = ["#90ee90"]
plt.bar(["6.2", "8.2", "4.3.1"], [correlation_6, correlation_8, correlation_431], color=custom_green_palette)
plt.title("Correlation of Different Price Measures Depending on Number of Contractors")
```
